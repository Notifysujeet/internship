#checking the chi square test

```python
import numpy as np
import pandas as pd 
import scipy.stats as stats
ds=pd.read_csv('role_website.csv')
ds
```

```python
df=pd.DataFrame(ds)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dtypes
```

#create contingency table

```python
df_crosstab=pd.crosstab(df['role'],df['website'],margins=True)
print(df_crosstab)
```

```python
df_crosstab.columns=['facebook','linkedin','row_totals']
df_crosstab.index=['student','teacher','col_totals']
df_crosstab
```

#get the observed table

```python
observed=df_crosstab.iloc[0:2,0:2]
observed
```

```python
 stats.chi2_contingency?
```

```python
 chi2,p,dof,expected=stats.chi2_contingency(observed=observed)
print('chi square test=',chi2)
print('p value=',p)
print('dof=',dof)
print('expected=',expected)
```

# pvalue less than significant value that is0.05 so null hypothesis is rejected

# ANOVA TEST=COMPARE THE MEANS OF MORE THAN 2 GROUPS

NULL HYPO=FROUUP MEANS ARE EAQUAL
ALT=AT LEAST ONE GORUP HAS DIFFERENT MEAN FROM OTHERS

```python
Marks_sectionA=[70,54,43,65,35,56]
Marks_sectionB=[34,44,55,66,33,45]
Marks_sectionc=[45,54,55,35,65,44]
stats.f_oneway(Marks_sectionA,Marks_sectionB,Marks_sectionc)
```

pvalue >0.05 it means null hypo is accepted that means all groups have similar means it means all students scored almost similar marks
